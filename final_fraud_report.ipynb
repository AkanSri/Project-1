```python
import pandas as pd
from pathlib import Path
import seaborn as sns
import hvplot.pandas
%matplotlib inline
```

## Fraud Activity Report Data Path Files.

```python
fraud_yr_2020= Path("FInCen CSV Data/FinCen SAR Stats 2020 (Months).csv")
fraud_yr_2020= pd.read_csv(fraud_yr_2020, thousands=',', index_col='Year Month')
fraud_yr_2020

fraud_yr_2019= Path("FInCen CSV Data/FinCen SAR Stats 2019 (Months).csv")
fraud_yr_2019= pd.read_csv(fraud_yr_2019, thousands=',', index_col='Year Month')
fraud_yr_2019

fraud_yr_2021= Path("FInCen CSV Data/FinCen SAR Stats 2021 (Months).csv")
fraud_yr_2021= pd.read_csv(fraud_yr_2021, thousands=',', index_col='Year Month')
fraud_yr_2021
```

## Remove rows with 'Totals' for each dataframe year. 

```python
fraud_yr_2020=fraud_yr_2020.loc[fraud_yr_2020["Industry"]!="[Total]" ]
fraud_yr_2020f=fraud_yr_2020.loc[fraud_yr_2020["Suspicious Activity"]!="[Total]" ]
fraud_yr_2020=fraud_yr_2020.loc[fraud_yr_2020["State"]!="[Total]" ]


fraud_yr_2021=fraud_yr_2021.loc[fraud_yr_2021["Industry"]!="[Total]" ]
fraud_yr_2021=fraud_yr_2021.loc[fraud_yr_2021["Suspicious Activity"]!="[Total]" ]
fraud_yr_2021=fraud_yr_2021.loc[fraud_yr_2021["State"]!="[Total]" ]


fraud_yr_2019=fraud_yr_2019.loc[fraud_yr_2019["Industry"]!="[Total]" ]
fraud_yr_2019f=fraud_yr_2019.loc[fraud_yr_2019["Suspicious Activity"]!="[Total]" ]
fraud_yr_2019=fraud_yr_2019.loc[fraud_yr_2019["State"]!="[Total]" ]
```

## Drop "Suspicious Activity" Colomn from Datatframes.

```python
fraud_2020_clean= fraud_yr_2020.drop(columns= 'Suspicious Activity')
fraud_2021_clean= fraud_yr_2021.drop(columns= 'Suspicious Activity')
fraud_2019_clean= fraud_yr_2019.drop(columns= 'Suspicious Activity')
fraud_2019_clean
```

## Combine All Fraud Reported into One Dataframe. 

```python
fraud_combined_data = pd.concat([fraud_2019_clean, fraud_2020_clean, fraud_2021_clean], axis=0, join="inner")
fraud_combined_data
```

## Filter combined dataframe by 'Depository Institution'

```python
fraud_combined_data = fraud_combined_data.loc[fraud_combined_data["Industry"] == 'Depository Institution']
fraud_combined_data
```

## Update combined data 'Date Time' format.

```python
fraud_combined_data = fraud_combined_data.reset_index()
fraud_combined_data['Year Month'] = pd.to_datetime(fraud_combined_data['Year Month'], infer_datetime_format = True)
fraud_combined_data['Count'] = fraud_combined_data['Count'].astype(float)

fraud_combined_data
```

## Group Combined Dataframes by 'State' and 'Count. 

```python
grouped_stats_state = fraud_combined_data[["State", "Count", "Year Month"]]

grouped_stats_state = grouped_stats_state.groupby(['Year Month', 'State']).sum()
grouped_stats_state
```

## Group fraud report data by 'State'.

```python
grouped_stats_state = grouped_stats_state.sort_index(level = 1, ascending = False)
grouped_stats_state
```

## Group Data by Year and Month

```python
grouped_stats_ym = fraud_combined_data.groupby(['Year Month']).sum()
grouped_stats_ym.head()
```

## Total Columns for each year. 

```python
grouped_stats_ym = grouped_stats_ym.rename(columns = {'Count' : 'Total'})
grouped_stats_ym.head()
```

## Disply the State with the Max number fraud reports. 

```python
pd.set_option("display.max_rows", 10, "display.max_columns", None)
grouped_stats_state = grouped_stats_state.rename(columns = {'Count' : 'Reports'})
grouped_stats_state
```

```python
pd.set_option("display.max_rows", 10, "display.max_columns", None)
all_fraud_data = pd.merge(grouped_stats_state, grouped_stats_ym, left_on = 'Year Month',right_index=True)
all_fraud_data
```

## Update fraud report data to percentage of fraud reports by State and Year Month

```python
all_fraud_data['Reporting Rate'] = round((all_fraud_data['Reports']/all_fraud_data['Total'])*100,3)
all_fraud_data
```

## Group total average of total fraud reports by state. 

```python
fraud_by_state = all_fraud_data.groupby('State').mean()
fraud_by_state = fraud_by_state.reset_index()
fraud_by_state.head()
```

```python
all_fraud_data['Year'] = all_fraud_data.index.get_level_values(0)
all_fraud_data['Year'] = all_fraud_data['Year'].dt.year
all_fraud_data
```

```python
max_data = all_fraud_data[['Reporting Rate','Year']].copy().reset_index()
max_data = max_data.drop(columns='Year Month')
max_data
max_data = max_data.groupby(['Year']).max()
max_data.index = max_data.index.map(str) + ' '+max_data['State']
```
