```python
# Imorting libraries 

import pandas as pd
from pathlib import Path
import seaborn as sns
import hvplot.pandas
%matplotlib inline

```

```python
# Setting up the path to the CSV file
csv_path = Path("../FInCen CSV Data/FinCen SAR Stats 2021 (Months).csv")

# Reading the csv file
csv_file = pd.read_csv(csv_path, index_col = "Year Month")
csv_file.head()
```

```python
#Using "loc" to remove the lines that have "Totals"

csv_file = csv_file.loc[csv_file["Industry"]!="[Total]" ]
csv_file = csv_file.loc[csv_file["Suspicious Activity"]!="[Total]" ]
csv_file = csv_file.loc[csv_file["State"]!="[Total]" ]
csv_file.head()
```

```python
grouped_stats = csv_file[["State", "Count"]]

grouped_stats = grouped_stats.groupby("State").count()
grouped_stats
```

```python

mapped_data = grouped_stats.hvplot.bar(rot=90,frame_width = 700, frame_height = 500, label = "FinCen Fraud SAR Reporting by State in 2021")
mapped_data
```

```python
csv_file.sort_values("Count")
mapped_data = csv_file.hvplot(y = "Count", groupby = "State", rot=90,frame_width = 700, frame_height = 500, label = "FinCen Fraud SAR Reporting by State in 2021")
mapped_data
```
